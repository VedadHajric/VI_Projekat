```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate, BatchNormalization, Activation, Add
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.applications import VGG19
from tensorflow.keras.applications.vgg19 import preprocess_input
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import load_model
from keras.saving import register_keras_serializable

import cv2 # OpenCV za učitavanje i manipulaciju slikama
import matplotlib.pyplot as plt

# --- Konfiguracija ---
IMG_WIDTH = 256
IMG_HEIGHT = 256
IMG_CHANNELS = 3 # RGB slike

# Putanje do podataka
BLURRED_IMAGES_PATH = '/content/drive/My Drive/ESPCN 64-_128/test_res'

SHARP_IMAGES_PATH = '/content/drive/My Drive/ESPCN 64-_128/test_hr2'




OUTPUT_DIR = 'rezultati_deblurringa/' # Direktorij za čuvanje modela i rezultata
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# Hiperparametri za treniranje
BATCH_SIZE = 8
EPOCHS = 100
LEARNING_RATE = 1e-4

# --- Pomoćne funkcije ---
def load_image(path, target_size=(IMG_HEIGHT, IMG_WIDTH)):
    """Učitava sliku, konvertuje u RGB, reskalira i normalizuje."""
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # OpenCV učitava kao BGR
    img = cv2.resize(img, target_size)
    img = (img / 255.0).astype(np.float32)# Normalizacija na [0, 1]
    return img

def get_image_paths(blurred_dir, sharp_dir):
    """Dobija putanje do slika i osigurava da su imena sortirana za uparivanje."""
    blurred_files = sorted([os.path.join(blurred_dir, fname) for fname in os.listdir(blurred_dir) if fname.lower().endswith(('.png', '.jpg', '.jpeg'))])
    sharp_files = sorted([os.path.join(sharp_dir, fname) for fname in os.listdir(sharp_dir) if fname.lower().endswith(('.png', '.jpg', '.jpeg'))])

    # Provjera da li imamo isti broj zamućenih i oštrih slika
    if len(blurred_files) != len(sharp_files):
        raise ValueError(f"Broj zamućenih ({len(blurred_files)}) i oštrih ({len(sharp_files)}) slika se ne poklapa. Provjerite direktorije.")

    # Dodatna provjera imena (osnovni dio imena treba da se poklapa)
    for b_path, s_path in zip(blurred_files, sharp_files):
        b_name = os.path.basename(b_path).split('.')[0].replace('_blurry','').replace('_blurred','').replace('_zamucena','')
        s_name = os.path.basename(s_path).split('.')[0].replace('_sharp','').replace('_ostra','')
        if b_name != s_name:
            print(f"UPOZORENJE: Moguće neusklađeno ime: {os.path.basename(b_path)} vs {os.path.basename(s_path)}")



    return blurred_files, sharp_files

def augment(blurred_img, sharp_img):
    """Primjenjuje iste nasumične transformacije na par slika."""
    # Nasumično okretanje lijevo-desno (50% šanse)
    if tf.random.uniform(()) > 0.5:
        blurred_img = tf.image.flip_left_right(blurred_img)
        sharp_img = tf.image.flip_left_right(sharp_img)

    # Nasumično okretanje gore-dole (50% šanse)
    if tf.random.uniform(()) > 0.5:
        blurred_img = tf.image.flip_up_down(blurred_img)
        sharp_img = tf.image.flip_up_down(sharp_img)

    # Nasumična rotacija za 90 stepeni (25% šanse za svaku rotaciju: 0, 90, 180, 270)
    # k je broj rotacija od 90 stepeni (0, 1, 2, ili 3)
    k = tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32)
    blurred_img = tf.image.rot90(blurred_img, k=k)
    sharp_img = tf.image.rot90(sharp_img, k=k)

    return blurred_img, sharp_img

def create_dataset(blurred_paths, sharp_paths, batch_size, augment_data=False, cache_data=False):
    """Kreira tf.data.Dataset za efikasno učitavanje."""
    def generator():
        for b_path, s_path in zip(blurred_paths, sharp_paths):
            blurred_img = load_image(b_path, (IMG_HEIGHT, IMG_WIDTH))
            sharp_img = load_image(s_path, (IMG_HEIGHT, IMG_WIDTH))
            yield blurred_img, sharp_img

    dataset = tf.data.Dataset.from_generator(
        generator,
        output_signature=(
            tf.TensorSpec(shape=(IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=tf.float32),
            tf.TensorSpec(shape=(IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=tf.float32)
        )
    )

    if augment_data:
        dataset = dataset.map(augment, num_parallel_calls=tf.data.AUTOTUNE)

    if augment_data:
        dataset = dataset.shuffle(buffer_size=len(blurred_paths))

    # Dodaj keširanje
    if cache_data:
        dataset = dataset.cache()

    dataset = dataset.batch(batch_size)
    dataset = dataset.repeat()
    dataset = dataset.prefetch(tf.data.AUTOTUNE)

    return dataset

# POMOĆNA FUNKCIJA: Konvolucioni blok
def conv_block(input_tensor, num_filters):
    """
    Standardni konvolucioni blok: Conv -> BN -> ReLU, ponovljeno dva puta.
    """
    # Prva konvolucija
    x = Conv2D(num_filters, (3, 3), padding='same', kernel_initializer='he_normal')(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Druga konvolucija
    x = Conv2D(num_filters, (3, 3), padding='same', kernel_initializer='he_normal')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    return x

# POMOĆNA FUNKCIJA: Attention blok
def attention_block(x, g, inter_channels):
    """
    Attention blok za U-Net.
    x: Skip connection tenzor iz enkodera.
    g: Tenzor iz dekodera.
    """
    theta_x = Conv2D(inter_channels, (1, 1), strides=(1, 1), padding='same', kernel_initializer='he_normal')(x)
    phi_g = Conv2D(inter_channels, (1, 1), strides=(1, 1), padding='same', kernel_initializer='he_normal')(g)

    f = Activation('relu')(Add()([theta_x, phi_g]))
    psi_f = Conv2D(1, (1, 1), strides=(1, 1), padding='same', kernel_initializer='he_normal')(f)

    rate = Activation('sigmoid')(psi_f)

    attn_x = layers.multiply([x, rate])
    return attn_x


# GLAVNA FUNKCIJA: Prepravljeni Attention U-Net
def build_attention_unet_v2(input_shape=(256, 256, 3)):
    """
    Gradi Attention U-Net model sa poboljšanjima:
    - Batch Normalization nakon svake konvolucije.
    - UpSampling2D + Conv2D umjesto Conv2DTranspose.
    - HeNormal inicijalizator težina.
    """
    inputs = Input(shape=input_shape)

    # --- ENKODER ---
    # Blok 1: 256x256 -> 128x128
    c1 = conv_block(inputs, 64)
    p1 = MaxPooling2D((2, 2))(c1)

    # Blok 2: 128x128 -> 64x64
    c2 = conv_block(p1, 128)
    p2 = MaxPooling2D((2, 2))(c2)

    # Blok 3: 64x64 -> 32x32
    c3 = conv_block(p2, 256)
    p3 = MaxPooling2D((2, 2))(c3)

    # Blok 4: 32x32 -> 16x16
    c4 = conv_block(p3, 512)
    p4 = MaxPooling2D((2, 2))(c4)

    # --- BOTTLENECK ---
    # Blok 5: 16x16
    c5 = conv_block(p4, 1024)

    # --- DEKODER ---
    # Blok 6: 16x16 -> 32x32
    u6 = UpSampling2D(size=(2, 2))(c5)
    # Ne treba nam konvolucija ovdje, jer attention block ima svoju
    att6 = attention_block(c4, u6, inter_channels=512)
    merge6 = concatenate([u6, att6])
    c6 = conv_block(merge6, 512)

    # Blok 7: 32x32 -> 64x64
    u7 = UpSampling2D(size=(2, 2))(c6)
    att7 = attention_block(c3, u7, inter_channels=256)
    merge7 = concatenate([u7, att7])
    c7 = conv_block(merge7, 256)

    # Blok 8: 64x64 -> 128x128
    u8 = UpSampling2D(size=(2, 2))(c7)
    att8 = attention_block(c2, u8, inter_channels=128)
    merge8 = concatenate([u8, att8])
    c8 = conv_block(merge8, 128)

    # Blok 9: 128x128 -> 256x256
    u9 = UpSampling2D(size=(2, 2))(c8)
    att9 = attention_block(c1, u9, inter_channels=64)
    merge9 = concatenate([u9, att9])
    c9 = conv_block(merge9, 64)

    # --- IZLAZNI SLOJ ---
    # 1x1 konvolucija za mapiranje na 3 kanala (RGB)
    # Sigmoid aktivacija za izlaz u opsegu [0, 1]
    outputs = Conv2D(3, (1, 1), activation='sigmoid')(c9)

    model = Model(inputs=[inputs], outputs=[outputs])
    return model

# Inicijalizuj VGG19 jednom izvan funkcije
vgg = VGG19(include_top=False, weights='imagenet', input_shape=(256, 256, 3))
vgg.trainable = False
selected_layers = ['block1_conv2', 'block2_conv2', 'block3_conv3']
vgg_outputs = [vgg.get_layer(name).output for name in selected_layers]
vgg_model = Model(inputs=vgg.input, outputs=vgg_outputs)

def perceptual_loss(y_true, y_pred):
    y_true = preprocess_input(y_true * 255.0)
    y_pred = preprocess_input(y_pred * 255.0)
    true_features = vgg_model(y_true)
    pred_features = vgg_model(y_pred)
    loss = tf.add_n([tf.reduce_mean(tf.square(t - p)) for t, p in zip(true_features, pred_features)])
    return loss


# --- Glavni dio skripte ---
if __name__ == '__main__':
    print("Učitavanje putanja do slika...")
    all_blurred_paths, all_sharp_paths = get_image_paths(BLURRED_IMAGES_PATH, SHARP_IMAGES_PATH)
    print(f"Pronađeno {len(all_blurred_paths)} parova slika.")

    if not all_blurred_paths:
        print("Nijedna slika nije pronađena. Provjerite putanje i ekstenzije slika.")
        exit()

    # Podjela na trening, validacioni i testni skup (70%, 20%, 10%)
    blurred_train_val, blurred_test, sharp_train_val, sharp_test = train_test_split(
        all_blurred_paths, all_sharp_paths, test_size=0.1, random_state=42, shuffle=True
    )
    relative_val_size = 0.2 / (1.0 - 0.1)
    blurred_train, blurred_val, sharp_train, sharp_val = train_test_split(
        blurred_train_val, sharp_train_val, test_size=relative_val_size, random_state=42, shuffle=True
    )

    print(f"Trening skup: {len(blurred_train)} slika")
    print(f"Validacioni skup: {len(blurred_val)} slika")
    print(f"Testni skup: {len(blurred_test)} slika")

    print("Kreiranje tf.data.Dataset objekata...")
    # Primijenite augmentaciju SAMO na trening skup
    train_dataset = create_dataset(blurred_train, sharp_train, BATCH_SIZE, augment_data=True,cache_data=False)
    # Bez augmentacije za validacioni skup
    val_dataset = create_dataset(blurred_val, sharp_val, BATCH_SIZE, augment_data=False,cache_data=True)

    # PSNR metrika
    def psnr(y_true, y_pred):
        return tf.image.psnr(y_true, y_pred, max_val=1.0)

    # SSIM metrika
    def ssim(y_true, y_pred):
        return tf.image.ssim(y_true, y_pred, max_val=1.0)

    # Vrijednosti koje treba podesiti
    ALPHA = 0.85 # Težina za SSIM
    BETA = 0.15 # Težina za L1
    GAMMA = 1e-6 # Težina za Perceptualni (istu koju ste već našli)

    @register_keras_serializable()
    def l1_ssim_perceptual_loss(y_true, y_pred):
        # L1 komponenta
        l1_loss = tf.reduce_mean(tf.abs(y_true - y_pred))

        # SSIM komponenta
        ssim_loss_val = 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))

        # Perceptualna komponenta
        perceptual_loss_val = perceptual_loss(y_true, y_pred)

        # Kombinovani gubitak
        total_loss = (ALPHA * ssim_loss_val) + \
                    (BETA * l1_loss) + \
                    (GAMMA * perceptual_loss_val)

        return total_loss

    print("Kreiranje Attention U-Net modela...")
    model = build_attention_unet_v2(input_shape=(IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS))
    model.compile(optimizer=Adam(learning_rate=LEARNING_RATE),  loss=l1_ssim_perceptual_loss, metrics=[psnr, ssim]) # Ovo je TensorFlow implementacija L1 gubitka
    model.summary()

    model_checkpoint_path = os.path.join(OUTPUT_DIR, 'unet_deblur_best_model.keras')
    callbacks = [
        EarlyStopping(patience=10, restore_best_weights=True, monitor='val_loss'),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5),
        ModelCheckpoint(filepath=model_checkpoint_path, save_best_only=True, monitor='val_loss')
    ]

    steps_per_epoch = len(blurred_train) // BATCH_SIZE
    validation_steps = len(blurred_val) // BATCH_SIZE

    for blurry_batch, sharp_batch in train_dataset.take(1):
      # L1 gubitak
      l1_val = tf.reduce_mean(tf.abs(sharp_batch - blurry_batch))

      # Perceptualni gubitak
      perceptual_val = perceptual_loss(sharp_batch, blurry_batch)

      print(f"Primjer vrijednosti na jednom batchu (Zamućena vs. Oštra):")
      print(f"L1 Gubitak: {l1_val.numpy()}")
      print(f"Perceptualni Gubitak: {perceptual_val.numpy()}")
      break

    if steps_per_epoch == 0 or validation_steps == 0:
        print("UPOZORENJE: Nema dovoljno podataka za barem jedan batch u treningu ili validaciji. Smanjite BATCH_SIZE ili dodajte više slika.")
    else:
        history = model.fit(
            train_dataset,
            epochs=EPOCHS,
            steps_per_epoch=steps_per_epoch,
            validation_data=val_dataset,
            validation_steps=validation_steps,
            callbacks=callbacks
        )

        print("Treniranje završeno.")
        model.save(os.path.join(OUTPUT_DIR, 'unet_deblur_final_model.keras'))

        # Prikaz historije treniranja
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 3, 1)
        plt.plot(history.history['loss'], label='Trening Loss')
        plt.plot(history.history['val_loss'], label='Validacioni Loss')
        plt.title('Loss')
        plt.xlabel('Epoha')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)

        plt.subplot(1, 3, 2)
        plt.plot(history.history['psnr'], label='Trening PSNR')
        plt.plot(history.history['val_psnr'], label='Validacioni PSNR')
        plt.title('PSNR')
        plt.xlabel('Epoha')
        plt.ylabel('PSNR (dB)')
        plt.legend()
        plt.grid(True)

        plt.subplot(1, 3, 3)
        plt.plot(history.history['ssim'], label='Trening SSIM')
        plt.plot(history.history['val_ssim'], label='Validacioni SSIM')
        plt.title('SSIM')
        plt.xlabel('Epoha')
        plt.ylabel('SSIM')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.savefig(os.path.join(OUTPUT_DIR, 'training_history.png'))
        plt.show()


    print("\n\n--- Evaluacija na testnom skupu ---")

    # Učitavanje najboljeg modela
    print("Učitavanje najboljeg sačuvanog modela...")
    best_model = load_model(
        model_checkpoint_path,
        custom_objects={
            'l1_ssim_perceptual_loss': l1_ssim_perceptual_loss,
            'psnr': psnr,
            'ssim': ssim,
            'attention_block': attention_block,
            'conv_block': conv_block
        }
    )

    # Učitavanje testnih slika u memoriju za evaluaciju
    test_blurred_images = np.array([load_image(p) for p in blurred_test])
    test_sharp_images = np.array([load_image(p) for p in sharp_test])

    if len(test_blurred_images) > 0:

        # Izračunavanje baseline metrika (Zamućena vs. Oštra) ---
        print("\n### POREĐENJE METRIKA NA CIJELOM TESTNOM SKUPU ###")

        # Konvertovanje u Tenzore za efikasno računanje
        blurred_tensor = tf.convert_to_tensor(test_blurred_images, dtype=tf.float32)
        sharp_tensor = tf.convert_to_tensor(test_sharp_images, dtype=tf.float32)

        # Računanje PSNR i SSIM za originalne zamućene slike
        baseline_psnr = tf.reduce_mean(psnr(sharp_tensor, blurred_tensor))
        baseline_ssim = tf.reduce_mean(ssim(sharp_tensor, blurred_tensor))

        print("\n1. Baseline Metrike (Originalna Zamućena vs. Oštra slika):")
        print(f"   - Prosječan PSNR: {baseline_psnr.numpy():.4f} dB")
        print(f"   - Prosječan SSIM: {baseline_ssim.numpy():.4f}")

        # Evaluacija modela na testnim podacima
        print("\n2. Metrike Modela (Deblurirana vs. Oštra slika):")
        test_loss, model_psnr, model_ssim = best_model.evaluate(test_blurred_images, test_sharp_images, verbose=0)
        print(f"   - Gubitak (Loss): {test_loss:.4f}")
        print(f"   - Prosječan PSNR: {model_psnr:.4f} dB")
        print(f"   - Prosječan SSIM: {model_ssim:.4f}")

        # Predikcija i prikaz nekoliko primjera sa testnog skupa
        print("\nGenerisanje vizuelnog poređenja za nekoliko primjera...")
        num_examples_to_show = min(5, len(test_blurred_images))

        if num_examples_to_show > 0:
            predicted_sharp_images = best_model.predict(test_blurred_images[:num_examples_to_show])

            plt.figure(figsize=(18, num_examples_to_show * 6))
            for i in range(num_examples_to_show):
                # Uzimanje pojedinačnih slika
                blurry_img = test_blurred_images[i]
                deblurred_img = predicted_sharp_images[i]
                sharp_img = test_sharp_images[i]

                # Računanje metrika za svaku sliku pojedinačno ---
                blurry_img_batch = np.expand_dims(blurry_img, axis=0)
                deblurred_img_batch = np.expand_dims(deblurred_img, axis=0)
                sharp_img_batch = np.expand_dims(sharp_img, axis=0)

                # Izračunavanje metrika za zamućenu sliku
                psnr_val_blurry = psnr(sharp_img_batch, blurry_img_batch).numpy()[0]
                ssim_val_blurry = ssim(sharp_img_batch, blurry_img_batch).numpy()[0]

                # Izračunavanje metrika za debluriranu sliku
                psnr_val_deblurred = psnr(sharp_img_batch, deblurred_img_batch).numpy()[0]
                ssim_val_deblurred = ssim(sharp_img_batch, deblurred_img_batch).numpy()[0]

                # Prikaz zamućene slike sa metrikama
                plt.subplot(num_examples_to_show, 3, i * 3 + 1)
                plt.imshow(blurry_img)
                plt.title(f"Zamućena (Test {i+1})\nPSNR: {psnr_val_blurry:.2f} dB, SSIM: {ssim_val_blurry:.2f}", fontsize=10)
                plt.axis('off')

                # Prikaz deblurirane slike sa metrikama
                plt.subplot(num_examples_to_show, 3, i * 3 + 2)
                plt.imshow(deblurred_img)
                plt.title(f"Deblurirana (Model)\nPSNR: {psnr_val_deblurred:.2f} dB, SSIM: {ssim_val_deblurred:.2f}", fontsize=10)
                plt.axis('off')

                # Prikaz originalne oštre slike
                plt.subplot(num_examples_to_show, 3, i * 3 + 3)
                plt.imshow(sharp_img)
                plt.title(f"Oštra (Ground Truth)", fontsize=10)
                plt.axis('off')

            plt.tight_layout(pad=3.0)
            plt.savefig(os.path.join(OUTPUT_DIR, 'test_predictions_comparison.png'))
            plt.show()
    else:
        print("Nema slika u testnom skupu za prikaz.")

    print(f"\nModel i rezultati sačuvani u direktoriju: {OUTPUT_DIR}")
```
